```python
#分块读取数据
i=0
loop =45 #
chunkSize = 100000 #指定读入的记录数
chunks = []
for i in range(loop):
    try:
        start = time.clock()
        chunk = reader.get_chunk(chunkSize) ## 从csv文件迭代读取
        chunks.append(chunk)
        end=time.clock()
        i=i+1
        print('第{}次用时{} 秒: 读入{} rows'.format(i,round((end - start),2), i * chunkSize))
        """if i>=2:
            chunks.append(chunk)
            end=time.clock()
            i=i+1
            print('第{}次用时{} 秒: 读入{} rows'.format(i,round((end - start),2), i * chunkSize))
        """
    except StopIteration:
        loop = False
        print("Iteration is stopped.")
df = pd.concat(chunks, ignore_index=True)
df.shape
```

```python
#用字典方式实现2015年按天统计 
from datetime import datetime
dict_years15={}
days=pd.date_range('2015-01',periods=365,freq='D')#DatetimeIndex序列  
#print(days)
for day in days.strftime("%Y-%m-%d"):  #取DatetimeIndex的”年-月-天”数据，%m十进制表示的月份
    today=df[day]     #按天查询数据
    #print(day)
    day_value=today.shape[0] #取每天的记录数
    #print(day_value)
    dict_years15.update({day:day_value}) #添加元素至字典中
#字典输出格式{'2015-01-01': 2283, '2015-01-02': 2591}
print(dict_years15) 
#dict_years15.keys()     #返回键
#dict_years15.values()   #返回值  
max(dict_years15,key=dict_years15.get) #返回最大值所对应的键
min(dict_years15,key=dict_years15.get) #返回最小值所对应的键
sorted(dict_years15.items(),key = lambda x:x[1],reverse = True)[0:10] #按字典的值进行排序，显示前10个元素
```

```python
#2015年数据绘图   
ts2015 = pd.Series(list(dict_years15.values()), index=pd.date_range('2015-02-01', periods=365))
ts2015[0:31].plot(grid=True, kind='line',figsize=(10, 3),title="January 2015")  
```

```python
#用字典方式实现2016年按天统计 
from datetime import datetime
dict_years16={}
days=pd.date_range('2016-01',periods=365,freq='D')#DatetimeIndex序列  
for day in days.strftime("%Y-%m-%d"):  #取DatetimeIndex的”年-月-天”数据，%m十进制表示的月份
    today=df[day]     #按天查询数据
    day_value=today.shape[0] #取每天的记录数
    dict_years16.update({day:day_value}) #添加元素至字典中
sorted(dict_years16.items(),key = lambda x:x[1],reverse = True)[0:10] #按字典的值进行排序，显示前10个元素
```

```python
#用字典方式实现2016-01-05按小时统计 
from datetime import datetime
dict_hours1={}
hours=pd.date_range(start='2016-01-05 00',periods=24,freq='H')#DatetimeIndex序列  

for day in hours.strftime("%Y-%m-%d %H"):  #取DatetimeIndex的”年-月-天”数据，%m十进制表示的月份
    hh=df[day]     #按天查询数据
    day_value=hh.shape[0] #取每天的记录数
    dict_hours1.update({day:day_value}) #添加元素至字典中
sorted(dict_hours1.items(),key = lambda x:x[1],reverse = True)[0:24] #按字典的值进行排序，显示前10个元素
```

```python
#用字典方式实现按时间序列的小时统计 
from datetime import datetime
dict_hours={}
dates=['2016-01-05 00','2016-01-19 00','2016-02-14 00','2016-12-16 00']#定义需要统计的日期列表
for d in range(len(dates)):
    hours=pd.date_range(start=dates[d],periods=24,freq='H')#DatetimeIndex序列  
    for day in hours.strftime("%Y-%m-%d %H"):  #取DatetimeIndex的”年-月-天”数据，%m十进制表示的月份
        hh=df[day]     #按天查询数据
        day_value=hh.shape[0] #取每天的记录数
        dict_hours.update({day:day_value}) #添加元素至字典中
#sorted(dict_hours.items(),key = lambda x:x[1],reverse = True)[0:-1] #按字典的值进行排序，显示前10个元素
#print(dict_hours)
```

```python
#基于时间序列的24h时数据折线图  
fig=plt.figure(figsize=(8,3)) #定义画布大小
hours=pd.date_range(start='2016-01-05 00',periods=24,freq='H')#DatetimeIndex序列 
#months=pd.date_range('2000-01',periods=12,freq='M')#生成DatetimeIndex序列 
x_ticks=list(hours.strftime("%H")) #时间间隔为小时单元
plt.title("24-Hours Data")
time_period=24

plt.plot(x_ticks,list(dict_hours.values())[0:time_period], color='green', label='2015-01-05',marker="o",linestyle="--")
plt.plot(x_ticks,list(dict_hours.values())[time_period:2*time_period], color='blue', label='2015-01-19',marker="*",linestyle="-.")
plt.plot(x_ticks,list(dict_hours.values())[2*time_period:3*time_period], color='red', label='2016-02-14',marker=".")
plt.plot(x_ticks,list(dict_hours.values())[3*time_period:4*time_period], color='y', label='2016-12-16',marker=".")

plt.legend()
plt.grid()
plt.show()
```
