```python
#【例3-1】操作系统版本信息
import wmi #导入wmi模块
c = wmi.WMI()  #初始化
os_info={}
for sys in c.CIM_OperatingSystem():
    #print(sys)
    os_info["OS_Name:"] = sys.Caption
    os_info["OS_Class Name:"] =sys.CreationClassName
    os_info["Manufacturer:"] =sys.Manufacturer
print(os_info)
```

```python
#【例3-2】中断信息
import wmi #导入wmi模块
c = wmi.WMI()  #实例化
for p in c.Win32_IRQResource(Name = "IRQ1"):
    print(p)
for p in c.Win32_IRQResource(Name = "IRQ8"):
    print(p)
```

```python
# 【例3-3】获取进程信息
import wmi
c = wmi.WMI ()
process_info={}
for process in c.Win32_Process (Name="python.exe"):
    process_info["Name"] = process.Name
    process_info["Process Id"] = process.ProcessId #进程标识符，PID唯一
    process_info["Session Id"] = process.SessionId #
    process_info["Priority"] = process.Priority
    process_info["ThreadCount"] = process.ThreadCount
    #print(process.ProcessId, process.Name)
    print(process)
print(process_info)
    
```

```python
#【例3-4】Python调用Win32_ Thread()方法获取线程信息
import wmi #导入wmi模块
c = wmi.WMI()
Thread_info={}
for thd in c.Win32_Thread(ProcessHandle = "4"): # 列出所有进程句柄为4的线程
    Thread_info["CreationClassName"]=thd.CreationClassName
    Thread_info["Process Handle"]=thd.ProcessHandle
    Thread_info["Start Address"]=thd.StartAddress
    Thread_info["Thread State"]=thd.ThreadState
    Thread_info["Handle"]=thd.Handle 
    #print(th)
print(Thread_info)
```

```python
#【例3-5】Python利用multiprocessing库实现多进程创建与管理
import multiprocessing
import os

def run_proc(name):
    print('Child process {0} {1} Running '.format(name, os.getpid())) 

if __name__ == '__main__':
    print('Parent process {0} is Running'.format(os.getpid())) 
    for i in range(3):
        p = multiprocessing.Process(target=run_proc, args=(str(i),))
        print('process start')
        p.start() #star() 方法启动进程
        p.join() #join()方法实现进程间的同步，等待所有进程退出。
        print('Process close')
```

```python
#【例3-6】Python调用Win32_Processor()方法获取Cache容量信息
import os,sys
import wmi #导入wmi模块
c = wmi.WMI()  #初始化
def Disp_Cache_info():
    Cache_Info = {} # 定义字典
    for cache in c.Win32_Processor():     
        Cache_Info["L2CacheSize:"] = cache.L2CacheSize    #二级缓存大小
        #Cache_Info["L2CacheSpeed"] = cache.L2CacheSpeed  #--二级缓存处理器的时钟速度
        Cache_Info["L3CacheSize:"] = cache.L3CacheSize    #三级缓存大小
        #Cache_Info["L3CacheSpeed"] = cache.L3CacheSpeed  #--二级缓存处理器的时钟速度
    return Cache_Info        
print(Disp_Cache_info())    #输出Cache信息  
    
```

```python
#【例3-7】Python调用Win32_OperatingSystem()方法获取虚拟存储容量信息
import os,sys
import wmi #导入wmi模块
c = wmi.WMI()  #实例化
v_mem={}
for vmem in c.Win32_OperatingSystem():
    v_mem["Total Virtual Memory Size:"]=int(vmem.TotalVirtualMemorySize)/(1024**2)
    v_mem["Total Visible MemorySize:"] =int(vmem.TotalVisibleMemorySize)/(1024**2)
    v_mem["Free Virtual Memory:"]=int(vmem.FreeVirtualMemory)/(1024**2)
print(v_mem)    
```

```python
# 【例3-9】Python显示指定路径下的所有子目录及文件
import os  
file_dir="C:\Program Files"
for root, dirs, files in os.walk(file_dir):  
    print(root) #当前目录路径  
    print(dirs) #当前路径下所有子目录  
    print(files) #当前路径下所有子目录子中的文件
```
